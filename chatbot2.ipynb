```python
import spacy
spacy.load('pt_core_news_sm')
from spacy.lang.pt import Portuguese
parser = Portuguese()
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
nltk.download('stopwords')
pt_stop = set(nltk.corpus.stopwords.words('portuguese'))
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in pt_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
text_2.to_csv('data/chatbot.csv', index=False)
```

```python
import random
text_data = []

#df = open('data/chatbot.csv')
df = open('data/FRASES.csv')

with df as f:
    for line in f:
        tokens = prepare_text_for_lda(line)
        #if random.random() > .99:
        print(tokens)
        text_data.append(tokens)
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]

import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
import gensim
NUM_TOPICS = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15, random_state=1)
ldamodel.save('model5.gensim')
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```
